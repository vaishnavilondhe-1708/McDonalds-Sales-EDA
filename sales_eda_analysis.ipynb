```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('menu.csv')
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
df = df.dropna(subset=['Calories'])
```

```python
plt.figure(figsize=(14, 8))
cat_avg = df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
plt.barh(cat_avg.index, cat_avg.values, color='#FF6B6B')
plt.title("McDonald's Menu Analysis\nAverage Calories by Category", fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Average Calories', fontsize=12)
plt.ylabel('Menu Category', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
```

```python
plt.savefig('category_analysis.png', dpi=300, facecolor='white', edgecolor='none', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
```

```python
plt.figure(figsize=(12, 8))
top10 = df.nlargest(10, 'Calories')
plt.barh(top10['Item'], top10['Calories'], color='#4ECDC4')
plt.title("Top 10 Highest Calorie Menu Items", fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Calories', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()

plt.savefig('top_items_analysis.png', dpi=300, facecolor='white', edgecolor='none', bbox_inches='tight')
plt.show()

print(" 2 PERFECT PNGs SAVED!")
print(" Left sidebar: category_analysis.png, top_items_analysis.png")
```
